```python
df = pd.read_csv('sales_data.csv')
```

```python
df
```

```python
product_revenue = df.groupby('ProductID')['Profit'].sum() #total revenue folr each product
product_revenue
```

```python
overall_revenue = df['Profit'].sum() #total revenue of overall
overall_revenue
```

```python
avg_profit_margin = df.groupby('ProductID')['Profit'].mean() / df.groupby('ProductID')['Price'].mean() #avg profit margin
avg_profit_margin
```

```python
quantity_sold = df.groupby('ProductID')['QuantitySold'].sum() #total quantity sold for each product
quantity_sold
```

```python
top_selling_products = quantity_sold.sort_values(ascending=False).head(5)
category_quantity_sold = df.groupby('Category')['QuantitySold'].sum()
top_selling_categories = category_quantity_sold.sort_values(ascending=False).head(5)
top_selling_categories
```
